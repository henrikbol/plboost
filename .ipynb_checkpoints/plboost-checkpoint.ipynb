```python
import pandas as pd
```

```python
cols = [
    "Season",
    "Date",
    "HomeTeam",
    "AwayTeam",
    "FTHG",
    "FTAG",
    "FTR",
    "HTHG",
    "HTAG",
    "HTR",
    "Referee",
    "HS",
    "AS",
    "HST",
    "AST",
    "HF",
    "AF",
    "HC",
    "AC",
    "HY",
    "AY",
    "HR",
    "AR",
    "MaxH",
    "MaxD",
    "MaxA",
    "AvgH",
    "AvgD",
    "AvgA"
]
```

```python
csv_files = {"E2020_21.csv" : "20-21",
             "E2019_20.csv" : "19-20",
             "E2018_19.csv" : "18-19",
             "E2017_18.csv" : "17-18",
             "E2016_17.csv" : "16-17",
            }

df = pd.DataFrame()

for file, season in csv_files.items():
    _results = pd.read_csv(file)
    _results["Season"] = season
    df = df.append(_results)
    
df = df[cols]
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```
